```python
import random
```

```python
fabrica1=[]
for i in range(100):
  numeroaleatorio=round(random.uniform(50, 100),2)
  fabrica1.append(numeroaleatorio)
print(fabrica1)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("Producción diaria de la Fabrica 1")
```
