```python
autos=pd.DataFrame({'Precio': autos['price']})
autos.describe()
```

```python
import math
n=len(autos)
K=int(1+ math.log2(n))
print("Valor de n=", n, "\nNúmero de intervalos=", K)
```

```python
pyplot.figure()
pyplot.hist(autos['Precio'], bins=K, edgecolor='black')
pyplot.title("Histograma de precios de automóviles")
pyplot.xlabel("Precio")
pyplot.ylabel("Frecuencia")
pyplot.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.boxplot(x='body-style', y='price', data=autos)

plt.title("Distribución del precio según tipo de vehículo ")
plt.xlabel("Tipo de vehículo")
plt.ylabel("Precio del automóvil")
plt.show()
```

```python
sns.boxplot(x="drive-wheels", y="price", data=autos)
plt.title("Distribución del precio según tipo de tracción")
plt.xlabel("Tipo de tracción (drive-wheels)")
plt.ylabel("Precio del automóvil")
plt.show()
```

```python
sns.boxplot(x="engine-location", y="price", data=autos)
plt.title("PRECIO Y UBICACIÓN DEL MOTOR")
plt.xlabel("Ubicación del motor")
plt.ylabel("Precio del automóvil")
plt.show()
```
