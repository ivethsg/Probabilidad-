```python
import matplotlib.pyplot as plt
import math
```

```python
x = [0,10,20,50]
p = [0.50,0.30,0.15,0.05]

print("Suma de probabilidades:", sum(p))

esperanza = 0
for i in range(len(x)):
    esperanza += x[i]*p[i]

esperanza_cuadrado = 0
for i in range(len(x)):
    esperanza_cuadrado += (x[i]**2)*p[i]

varianza = esperanza_cuadrado - esperanza**2
desviacion = math.sqrt(varianza)

print(f"E(X)= {esperanza:.2f}")
print(f"E(X^2)= {esperanza_cuadrado:.2f}")
print(f"Varianza= {varianza:.2f}")
print(f"Desviación estándar= {desviacion:.2f}")

F = []
acum = 0
for prob in p:
    acum += prob
    F.append(acum)

for xi, fi in zip(x, F):
    print(f"F({xi})= {fi:.2f}")

plt.stem(x, p)
plt.title("Distribución Ejercicio 2")
plt.xlabel("X")
plt.ylabel("P(X)")
plt.grid(True)
plt.show()

plt.step(x, F, where='post')
plt.title("FDA Ejercicio 2")
plt.xlabel("X")
plt.ylabel("Probabilidad acumulada")
plt.grid(True)
plt.show()
```
