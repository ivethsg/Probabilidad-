```python
import matplotlib.pyplot as plt
```

```python
x=[5,6,7,8,9,10]
p=[0.05,0.10,0.20,0.30,0.25,0.10]

valor_esperado=0

for i in range(len(x)):
  xi=x[i]
  pi=p[i]
  producto=xi*pi
  valor_esperado+=producto
  print(f"x={xi},p={pi},x*p={producto:.2f}")

print(f"\nValor esperado(Esperanza):{valor_esperado:.2f}")
```
