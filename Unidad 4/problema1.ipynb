```python
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt
```

```python
media_personas = 5

prob_exactamente_2 = poisson.pmf(2, media_personas)
print("a). Probabilidad de que el número de personas sea dos:", prob_exactamente_2)

prob_exceda_10 = 1 - poisson.cdf(10, media_personas)
print("b). Probabilidad de que x exceda de 10:", prob_exceda_10)
print("Explicación: No es muy probable, ya que la probabilidad es apenas del", round(prob_exceda_10 * 100, 2), "%")

valores_x = np.arange(0, 16)
probabilidades_x = [poisson.pmf(x, media_personas) for x in valores_x]

plt.figure(figsize=(8,6))
plt.bar(valores_x, probabilidades_x, color='skyblue')
plt.title('Distribución de Poisson (media=5)')
plt.xlabel('Número de personas (x)')
plt.ylabel('Probabilidad')
plt.xticks(valores_x)
plt.grid(True)
plt.show()
```
