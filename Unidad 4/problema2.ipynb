```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
```

```python
n = 1000
p = 0.001

prob_0_defectuosos = binom.pmf(0, n, p)
prob_3_defectuosos = binom.pmf(3, n, p)
prob_4_defectuosos = binom.pmf(4, n, p)

print("Resultados del embarque de 1,000 motores:")
print(f"Probabilidad de que ninguno sea defectuoso: {prob_0_defectuosos}")
print(f"Probabilidad de que tres sean defectuosos: {prob_3_defectuosos}")
print(f"Probabilidad de que cuatro sean defectuosos: {prob_4_defectuosos}")

valores_x = np.arange(0, 11)
probabilidades_x = [binom.pmf(x, n, p) for x in valores_x]

print("\nValores de x\t Probabilidades")
for x, prob in zip(valores_x, probabilidades_x):
    print(f"{x}\t\t{prob}")

plt.figure(figsize=(8,6))
plt.bar(valores_x, probabilidades_x, color='pink')
plt.title('Distribución Binomial (n=1000, p=0.001)')
plt.xlabel('Número de motores defectuosos (x)')
plt.ylabel('Probabilidad')
plt.xticks(valores_x)
plt.grid(True)
plt.show()
```
